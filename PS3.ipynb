```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import special
```

```python
tau = [.01,.1,.5,1,2,5]
x = np.linspace(.0001, 10, 1000)

plt.vlines(1, 0, 1e3, color='blueviolet', label=r'$\tau$=0', lw=2)

colors = ['dodgerblue', 'lightseagreen', 'forestgreen', 'orange', 'orangered', 'deeppink']
for t, c in zip(tau, colors): 
    sig_t = (t * x**0.25)**-1 * np.exp(-(1+x**2)/t) * special.iv(0.25, 2*x/t)
    plt.plot(x, sig_t, label=r'$\tau$='+str(t), color=c, lw=2)
    #print(sig_t)

plt.title('Viscous Diffusion of an Accretion Disk')
plt.xlabel('r/R0')
plt.ylabel(r'$\Sigma \pi R_{0}^{2}$/m')
plt.ylim(0,3.5)
plt.xlim(.005, 10)
plt.xscale('log')
plt.legend(loc='upper left')
```
